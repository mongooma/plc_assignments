```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# clock() version 1 
t = pd.read_csv("./logs/log_cloud_4am.txt", header=1)
t1 = pd.read_csv("./logs/log_cloud_ripple_4am.txt", header=0)
# clock() version 2 
t2 = pd.read_csv("./logs/log_cloud.txt", header=1)
t3 = pd.read_csv("./logs/log_cloud_ripple.txt", header=0)
```

```python
# clock() version 1 

# plot 1: First, plot the execution time of 2, 4, 8, 16 and 32 rank runs as function of their number of ranks
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

fig.delaxes(axes[-1][-1])
fig.delaxes(axes[-1][-2])

i = 0
for test in range(1, 8):
    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array(t[t["rank"] != 1][t.test==test][t.barrier==1].groupby(["rank"])["time"].mean()), color='darksalmon', label="with barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array(t[t["rank"] != 1][t.test==test][t.barrier==1].groupby(["rank"])["time"].mean()), color='darksalmon')
    
    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array(t[t["rank"] != 1][t.test==test][t.barrier==0].groupby(["rank"])["time"].mean()), color='firebrick', label="without barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array(t[t["rank"] != 1][t.test==test][t.barrier==0].groupby(["rank"])["time"].mean()), color='firebrick')

    axes[i//3][i%3].set_title("test=%s" % test)
    axes[i//3][i%3].set_xticks([2, 4, 8, 16, 32])
    axes[i//3][i%3].set_xticklabels(['2', '4', '8', '16', '32'])
    axes[i//3][i%3].set_xlabel("rank")
    axes[i//3][i%3].set_ylabel("time(seconds)")
    axes[i//3][i%3].legend()

    i += 1

plt.tight_layout()
fig.suptitle("Rank-time", verticalalignment="bottom", y=0)
fig.savefig("rank_time.png", dpi=300)

# clock() version 2

# plot 1: First, plot the execution time of 2, 4, 8, 16 and 32 rank runs as function of their number of ranks
fig, axes = plt.subplots(3, 3, figsize=(10, 10))

fig.delaxes(axes[-1][-1])
fig.delaxes(axes[-1][-2])

i = 0
for test in range(1, 8):
    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array(t2[t2["rank"] != 1][t2.test==test][t2.barrier==1].groupby(["rank"])["time"].mean()), color='darksalmon', label="with barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array(t2[t2["rank"] != 1][t2.test==test][t2.barrier==1].groupby(["rank"])["time"].mean()), color='darksalmon')
    
    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array(t2[t2["rank"] != 1][t2.test==test][t2.barrier==0].groupby(["rank"])["time"].mean()), color='firebrick', label="without barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array(t2[t2["rank"] != 1][t2.test==test][t2.barrier==0].groupby(["rank"])["time"].mean()), color='firebrick')

    axes[i//3][i%3].set_title("test=%s" % test)
    axes[i//3][i%3].set_xticks([2, 4, 8, 16, 32])
    axes[i//3][i%3].set_xticklabels(['2', '4', '8', '16', '32'])
    axes[i//3][i%3].set_xlabel("rank")
    axes[i//3][i%3].set_ylabel("time(seconds)")
    axes[i//3][i%3].legend()

    i += 1

plt.tight_layout()
fig.suptitle("Rank-time", verticalalignment="bottom", y=0)
fig.savefig("rank_time_2.png", dpi=300)
```

```python
# clock() version 1

# plot 2: Next, plot of the speedup relative to the execution time of the serial MPI CLA adder of the 2, 4, 8, 16 and 32 rank runs

fig, axes = plt.subplots(3, 3, figsize=(10, 10))

fig.delaxes(axes[-1][-1])
fig.delaxes(axes[-1][-2])

i = 0
for test in range(1, 8):
    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array((t1[t1.test==test]["time"].mean() - 
                                                        t[t["rank"] != 1][t.test==test][t.barrier==1].groupby(["rank"])["time"].mean()) / 
                                                           float(t1[t1.test==test]["time"].mean())
                                                            ), color='darksalmon', label="with barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array((t1[t1.test==test]["time"].mean() - 
                                                        t[t["rank"] != 1][t.test==test][t.barrier==1].groupby(["rank"])["time"].mean()) / 
                                                           float(t1[t1.test==test]["time"].mean())
                                                         ), color='darksalmon')

    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array((t1[t1.test==test]["time"].mean() - 
                                                        t[t["rank"] != 1][t.test==test][t.barrier==0].groupby(["rank"])["time"].mean()) / 
                                                           float(t1[t1.test==test]["time"].mean())
                                                            ), color='firebrick', label="without barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array((t1[t1.test==test]["time"].mean() - 
                                                        t[t["rank"] != 1][t.test==test][t.barrier==0].groupby(["rank"])["time"].mean()) / 
                                                           float(t1[t1.test==test]["time"].mean())
                                                         ), color='firebrick')

    axes[i//3][i%3].set_title("test=%s" % test)
    axes[i//3][i%3].set_xticks([2, 4, 8, 16, 32])
    axes[i//3][i%3].set_xticklabels(['2', '4', '8', '16', '32'])
    axes[i//3][i%3].set_xlabel("rank")
    axes[i//3][i%3].set_ylabel("speedup( to ripple adder)")
    axes[i//3][i%3].legend()

    i += 1

plt.tight_layout()

fig.suptitle("Speed-up vs Ripple Adder", verticalalignment="bottom", y=0)

fig.savefig("Speed-up vs Ripple Adder.png", dpi=300)


# clock() version 2

# plot 2: Next, plot of the speedup relative to the execution time of the serial MPI CLA adder of the 2, 4, 8, 16 and 32 rank runs

fig, axes = plt.subplots(3, 3, figsize=(10, 10))

fig.delaxes(axes[-1][-1])
fig.delaxes(axes[-1][-2])

i = 0
for test in range(1, 8):
    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array((t3[t3.test==test]["time"].mean() - 
                                                        t2[t2["rank"] != 1][t2.test==test][t2.barrier==1].groupby(["rank"])["time"].mean()) / 
                                                           float(t3[t3.test==test]["time"].mean())
                                                            ), color='darksalmon', label="with barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array((t3[t3.test==test]["time"].mean() - 
                                                        t2[t2["rank"] != 1][t2.test==test][t2.barrier==1].groupby(["rank"])["time"].mean()) / 
                                                           float(t3[t3.test==test]["time"].mean())
                                                         ), color='darksalmon')

    axes[i//3][i%3].scatter([2, 4, 8, 16, 32], np.array((t3[t3.test==test]["time"].mean() - 
                                                        t2[t2["rank"] != 1][t2.test==test][t2.barrier==0].groupby(["rank"])["time"].mean()) / 
                                                           float(t3[t3.test==test]["time"].mean())
                                                            ), color='firebrick', label="without barrier")
    axes[i//3][i%3].plot([2, 4, 8, 16, 32], np.array((t3[t3.test==test]["time"].mean() - 
                                                        t2[t2["rank"] != 1][t2.test==test][t2.barrier==0].groupby(["rank"])["time"].mean()) / 
                                                           float(t3[t3.test==test]["time"].mean())
                                                         ), color='firebrick')

    axes[i//3][i%3].set_title("test=%s" % test)
    axes[i//3][i%3].set_xticks([2, 4, 8, 16, 32])
    axes[i//3][i%3].set_xticklabels(['2', '4', '8', '16', '32'])
    axes[i//3][i%3].set_xlabel("rank")
    axes[i//3][i%3].set_ylabel("speedup( to ripple adder)")
    axes[i//3][i%3].legend()

    i += 1

plt.tight_layout()

fig.suptitle("Speed-up vs Ripple Adder", verticalalignment="bottom", y=0)

fig.savefig("Speed-up vs Ripple Adder_2.png", dpi=300)

```
